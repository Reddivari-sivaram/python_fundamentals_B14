```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt  # importing a separete module from this librabry 
```

```python
# To give graphical presentation of the data
```

```python
x=[1,2,3]
y=[1,4,9]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.title("sales indication")
plt.xlabel("sales per year")
plt.ylabel("revenue per year")
plt.plot(x,y)
plt.show()
```
